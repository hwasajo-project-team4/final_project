```python
goods_list = df2['상품명'].unique()
goods_list
```

```python
train_df = pd.DataFrame(goods_list,columns=['상품명'])
train_df
```

```python
user_info_list = []
for goods in goods_list:
    age = df2[df2["상품명"] == goods]['나이'].value_counts().index[0]
    gender = df2[df2["상품명"] == goods]['성별'].value_counts().index[0]
    skin_type = df2[df2["상품명"] == goods]['피부타입'].value_counts().index[0]
    skin_trouble = df2[df2["상품명"] == goods]['피부트러블'].value_counts().index[0]
    user_info = age + ',' + gender + ',' + skin_type + ',' + skin_trouble
    user_info_list.append(user_info)
user_info_list
```

```python
train_df['사용자 정보'] = user_info_list
train_df
```

```python
brand_list = []
for goods in goods_list:
    brand = df2[df2['상품명'] == goods]['브랜드명'].unique()[0]
    brand_list.append(brand)
train_df['브랜드명'] = brand_list
train_df
```

```python
cols = ['브랜드명', '상품명', '사용자 정보']
train_df = train_df[cols]
train_df
```

```python
train_df.to_csv('스킨케어_학습데이터.csv')
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
count_vect = CountVectorizer(min_df=0, ngram_range=(1,2))
user_info_mat = count_vect.fit_transform(train_df['사용자 정보'])
user_info_mat.shape
```

```python
test = ['20대,남성,복합성,탄력없음']
test_mat = count_vect.transform(test)
test_mat.shape
```

```python
test_sim = cosine_similarity(test_mat, user_info_mat)
test_sim
```

```python
recommend_index_list = test_sim[0].argsort()[-10:][::-1]
recommend_index_list
```

```python
train_df.iloc[recommend_index_list]
```

```python
def find_sim_cosmetics(df, count_vect, user_info_mat, user_info, top_n=10):
    mat = count_vect.transform(user_info)
    user_info_sim = cosine_similarity(mat, user_info_mat)
    recommend_index_list = user_info_sim[0].argsort()[-top_n:][::-1]
    return df.iloc[recommend_index_list]
```

```python
test = ['20대,남성,건성,건조함']
top_n = 10
find_sim_cosmetics(train_df,count_vect,user_info_mat,test,top_n)
```
